```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
adam_lr01 = pd.read_csv("C:/Users/met48/Desktop/ABM-Surrogate/transformer_adam_lr01_epstein_loss.csv")
adam_lr001 = pd.read_csv("C:/Users/met48/Desktop/ABM-Surrogate/transformer_adam_lr001_epstein_loss.csv")
adam_lr0001 = pd.read_csv("C:/Users/met48/Desktop/ABM-Surrogate/transformer_adam_lr0001_epstein_loss.csv")

adamW_lr01 = pd.read_csv("C:/Users/met48/Desktop/ABM-Surrogate/transformer_adamW_lr01_epstein_loss.csv")
adamW_lr001 = pd.read_csv("C:/Users/met48/Desktop/ABM-Surrogate/transformer_adamW_lr001_epstein_loss.csv")
adamW_lr0001 = pd.read_csv("C:/Users/met48/Desktop/ABM-Surrogate/transformer_adamW_lr0001_epstein_loss.csv")
```

```python
fig, ax = plt.subplots()
ax.set_yscale('log')
ax.set_xticks([0, 5, 10, 15, 20])
plt.plot(adam_lr01, label="Adam LR 0.01")
plt.plot(adam_lr001, label="Adam LR 0.001")
plt.plot(adam_lr0001, label="Adam LR 0.0001")
plt.plot(adamW_lr01, label="AdamW LR 0.01")
plt.plot(adamW_lr001, label="AdamW LR 0.001")
plt.plot(adamW_lr0001, label="AdamW LR 0.0001")
plt.legend()
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.show()
```
